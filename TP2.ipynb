## Question 1
- Si au moins 1 des successeurs a au moins 1 valeur négative , la configuration du joueur actuel est gagnante. Il faudra choisir le maximum des valeurs negatives du successeur.
Vnext =  1 +  max (valeurs négatives des successeurs)

- Si tout les successeurs on une valeur positives, toute les configurations sont perdantes. Et donc il prendra le max des successeurs
Vnext = - 1 - max(valeurs des successeurs)

- Si aucun coup n'est possible V_act = 0

```python
def get_best_succ (l):
    neg = [x for x in l if x < 0]
    if neg == [] :
        return - max(l) - 1
    else :
        return abs(max(neg)) + 1

print(get_best_succ([2 , 6 , 11]))
print(get_best_succ([-11 , 3 , -8, 17]))
```
